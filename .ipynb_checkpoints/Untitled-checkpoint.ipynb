```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

pd.set_option('display.max_columns', 500)
```

```python
data_2018_raw = pd.read_csv('data\PlayerGame.2018.csv')
```

```python
data_2018_raw
```

```python
data_2018_raw.info()
```

```python
data_2018 = data_2018_raw.copy()
```

```python
data_2018['FantasyPosition'].value_counts()
```

```python
# Only keep rows that are needed
data_2018 = data_2018[(data_2018['SeasonType'] == 1) & \
                      (data_2018['PositionCategory'].isin(['OFF', 'ST'])) & \
                      (data_2018['FantasyPosition'].isin(['WR', 'RB', 'TE', 'QB', 'K']))] 

data_2018.head()
```

```python
groupby_name_cols = ['Name', 'Season','Position']
```

```python
data_2018.groupby(groupby_name_cols)['FantasyPoints'].agg(['mean','median']).reset_index().sort_values(by='mean', ascending=False)
```

```python
def top_20_median_score(position):
    groupby_name_cols = ['Name', 'Season','Position']
    top_20_median_score = pd.DataFrame(data_2018.groupby(groupby_name_cols)['FantasyPoints'].agg(['mean','median']).reset_index().sort_values(by='mean', ascending=False))
    top_20_median_score = top_20_median_score[top_20_median_score['Position'] == position].head(20)
    return top_20_median_score
```
