```python
data_2018 = data_2018_raw.copy()
```

```python
data_2018['FantasyPosition'].value_counts()
```

```python
# Only keep rows that are needed
data_2018 = data_2018[(data_2018['SeasonType'] == 1) & \
                      (data_2018['PositionCategory'].isin(['OFF', 'ST'])) & \
                      (data_2018['FantasyPosition'].isin(['WR', 'RB', 'TE', 'QB', 'K']))] 

data_2018.head()
```

```python
def top_scorers(position=None, num_results=20):
    groupby_name_cols = ['Name', 'Season','Position']
    keeper_players = ['Kenny Golladay','Marlon Mack','Nick Chubb','Christian Kirk', 'Aaron Rodgers', 'Josh Allen',
                      'JuJu Smith-Schuster', 'Sony Michel', 'Aaron Jones', 'Patrick Mahomes', 'Alvin Kamara',
                      'Adrian Peterson', 'Chris Godwin', 'Eric Ebron', 'Carson Wentz', 'Calvin Ridley', 'Josh Reynolds',
                      'James Connor', 'George Kittle', 'D.J. Moore', 'Dante Pettis']
    top_scorers = pd.DataFrame(data_2018.groupby(groupby_name_cols)['FantasyPoints'].agg(['mean','median']).reset_index().sort_values(by='mean', ascending=False))
    top_scorers = top_scorers[~top_scorers['Name'].isin(keeper_players)]
    
    if position == None:
        top_scorers = top_scorers.head(num_results)
    else:
        top_scorers = top_scorers[top_scorers['Position'].isin(position)].head(num_results)
    return top_scorers
```

```python
top_scorers(['RB'], 12)
```

```python
top_scorers_weekly = data_2018[data_2018['Name'].isin(top_scorers(['RB'], 12)['Name'])]
```

```python
sns.lineplot(x='Week', y='FantasyPoints', hue='Name',
             data=top_scorers_weekly)
```
